## 1. Load the saved embeddings

```python
# imports
import pandas as pd
import numpy as np
import random

random.seed(47)
```

```python
input_datapath = "data/english_embeddings_v1.csv"  # to save space, we provide a pre-filtered dataset
df = pd.read_csv(input_datapath, index_col=0)
```

```python
print(df.shape)
df.head()
```

```python
categories = sorted(df["intent"].unique())
print("Categories of samples:\n\n", df["intent"].value_counts())
```

```python
# embedding_matrix = df.embedding.tolist()
embedding_matrix = np.array(df.embedding.apply(eval).to_list())
embedding_matrix[0]
```

## 2. Reduce the embedding dimensionality to 2D

### 2.1 PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_dims = pca.fit_transform(embedding_matrix)
df["pca_embeddings"] = pca_dims.tolist()
df.head(2)
```

### 2.2 TSNE

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=15, random_state=42, init='random', learning_rate=200)
tsne_dims = tsne.fit_transform(embedding_matrix)
tsne_dims.shape
```

```python
df["tsne_embeddings"] = tsne_dims.tolist()
df.head(2)
```

## 3. Plot the embeddings of lower dimensionality in 2D

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

### 3.1 TSNE embeddings

```python
x = [x for x,y in tsne_dims]
y = [y for x,y in tsne_dims]
plt.figure(figsize=(12,6))

# classes = pd.Categorical(df.intent).codes # convert labels into array of integers
# scatter = plt.scatter(x, y, c=classes)
# plt.legend(handles=scatter.legend_elements()[0], labels=categories)
# plt.show()

data = pd.DataFrame({"X": x, "Y": y, "Category": df.intent})
print(data.head())
# Plot data by category
groups = data.groupby("Category")
for name, group in groups:
    plt.plot(group["X"], group["Y"], marker="o", linestyle="", label=name)
plt.show()
plt.legend()
```

### 3.2 PCA embeddings

```python
x = [x*10 for x,y in pca_dims]
y = [y*10 for x,y in pca_dims]
plt.figure(figsize=(12,6))

# classes = pd.Categorical(df.intent).codes # convert labels into array of integers
# scatter = plt.scatter(x, y, c=classes)
# plt.legend(handles=scatter.legend_elements()[0], labels=categories)
# plt.show()

data = pd.DataFrame({"X": x, "Y": y, "Category": df.intent})
print(data.head())
# Plot data by category
groups = data.groupby("Category")
for name, group in groups:
    plt.plot(group["X"], group["Y"], marker="o", linestyle="", label=name)
plt.show()
plt.legend()
```

## 4. Plot the embeddings of lower dimensionality in 3D

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca_dims = pca.fit_transform(embedding_matrix)
df["pca_3d_embeddings"] = pca_dims.tolist()
df.head(2)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
cmap = plt.get_cmap("tab20")

# Plot each sample category individually such that we can set label name.
for i, cat in enumerate(categories):
    sub_matrix = np.array(df[df["intent"] == cat]["pca_3d_embeddings"].to_list())
    x=sub_matrix[:, 0]
    y=sub_matrix[:, 1]
    z=sub_matrix[:, 2]
    colors = [cmap(i/len(categories))] * len(sub_matrix)
    ax.scatter(x*10, y*10, zs=z, zdir='z', c=colors, label=cat)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend(bbox_to_anchor=(1.1, 1))
```
